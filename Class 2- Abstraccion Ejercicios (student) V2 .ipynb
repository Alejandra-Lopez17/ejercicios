```python
class Robot:
    def __init__(self, name, color):
        self.name = name
        self.color = color
        self.age = 1
        self.batery = 8.5
        self.on = True
    
    def greet(self):
        print("hola mi nombre es: ", self.name)
        print("Mi edad es: ", self.age)
        print("Mi porcentaje de bateria es: ", self.batery)
        print("Mi estado de encendido es: ", self.on)


jarvis = Robot("jarvis", "rojo")
jarvis.greet()

ultron = Robot("ultron", "verde")
ultron.greet()

robotina = Robot("robotina", "azul")
robotina.greet()
```

### Ejercicio 3
A continuacion puedes una imagen con muchas flores:
image.png

te pedimos definir una clase `Flor` donde saques todos los atributos y metodos necesarios que definan a todas las flores

```python
class Flower:
    def __init__(self, nombre, color, altura, olor):
        self.nombre = nombre
        self.color = color
        self.altura = altura
        self.olor = olor
        
    def crecer(self, incremento):
        self.altura += incremento
        
    def cambiar_color(self, nuevo_color):
        self.color = nuevo_color
        
    def cambiar_olor(self, nuevo_olor):
        self.olor = nuevo_olor
        
    def mostrar_informacion(self):
        print("Nombre: ", self.nombre)
        print("Color: ", self.color)
        print("Altura: ", self.altura)
        print("Olor: ", self.olor)

Flower = Flower("Rosa", "Rojo", 20, "Agradable")
Flower.mostrar_informacion()
Flower.crecer(5)
Flower.cambiar_color("Blanco")
Flower.cambiar_olor("Fuerte")
Flower.mostrar_informacion()

```

### Ejercicio 4

Una parte muy importante en la abstraccion es poder distinguir entre los atributos que necesitamos y aquellos que podemos omitir o que no son relevantes para el problema que queremos resolver.

Intentemos resolver el siguiente problema, la empresa de Transporte Cotoca, encargada de viajes entre departamentos,  te contrato para que empecemos a desarrolar un sistema que los ayude con la venta de pasajes, para ello la primera cosa que necesitamos abstraer es un pasajero (que es la persona quien compra el pasaje para realizar un viaje), asi que te pidieron que abstraigas las propiedades y metodos necesarios de un pasajero


```python
class Passenger:
    def __init__(self, nombre, id, telefono):
        self.nombre = nombre
        self.id = id
        self.telefono = telefono

    def informacion(self):
        print(self.nombre)
        print(self.id)
        print(self.telefono)

pasajero = Passenger("Alejandra", 205489599, 31548967210)
pasajero.informacion()
```

```python
class Car:
    def __init__(self, model, brand, color):
        self.model = model
        self.brand = brand
        self.color = color

    def __str__(self):
        return f"Modelo: {self.model}\nMarca: {self.brand}\nColor: {self.color}"

    def run(self):
        print(f"El auto {self.model} de la marca {self.brand} y color {self.color} est en movimiento.")

toyota = Car('Auris', 'Toyota', 'Cafe')
print(toyota)
toyota.run()
```
